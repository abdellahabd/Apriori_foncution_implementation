```python
import pandas as pd 
# Sample dataset
data = {
    'age': [25, 35, 45, 55, 65,10],
    'income': [40000, 60000, 80000, 100000, 120000,120000]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
# Define categories for age
age_bins = [20, 30, 40, 50, 60, 70]
age_labels = ['20-29', '30-39', '40-49', '50-59', '60-69']

# Define categories for income
income_bins = [30000, 50000, 70000, 90000, 110000, 130000]
income_labels = ['30k-49k', '50k-69k', '70k-89k', '90k-109k', '110k-129k']

# Convert numerical features to nominal features
df['age_category'] = pd.cut(df['age'], bins=age_bins, labels=age_labels)
df['income_category'] = pd.cut(df['income'], bins=income_bins, labels=income_labels)

# Drop the original numerical features if needed
df.drop([ 'age','income'], axis=1, inplace=True)

df.head()
```

```python
d=pd.read_csv("./Medicaid1986.csv")
d.drop(["health1","health2"], axis=1, inplace=True)
d.head()
```

```python
bins = {
    'visits': [-1, 1, 3, 6, 10, float('inf')],
    'exposure': [0, 30, 60, 90, 120, float('inf')],
    'children': [-1, 0, 1, 2, 3, float('inf')],
    'age': [0, 18, 30, 50, 70, float('inf')],
    'income': [-1, 30, 50, 70, 90, float('inf')],
    'access': [ 0, 1, float('inf')],

}

labels = {
    'visits': ['0-1', '2-3', '4-6', '7-10', '10+'],
    'exposure': ['0-30', '31-60', '61-90', '91-120', '120+'],
    'children': ['None', 'One', 'Two', 'Three', 'Four or more'],
    'age': ['Child', 'Young Adult', 'Adult', 'Middle-Aged', 'Senior'],
    'income': ['Low', 'Lower Middle', 'Middle', 'Upper Middle', 'High'],
    'access': ['Low Access', 'High Access'],

}

for feature in bins.keys():
    d[feature + '_category'] = pd.cut(d[feature], bins=bins[feature], labels=labels[feature])

d.drop(list(bins.keys()), axis=1, inplace=True)
```
